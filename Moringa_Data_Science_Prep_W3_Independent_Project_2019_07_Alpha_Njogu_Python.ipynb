```
# IMPORTING THE PANDAS LIBRARY.
import pandas as pd
# IMPORTING THE NUMPYS LIBRARY
import numpy as np
```

```
# CREATING A DATAFRAME FOR Telcom_dataset1(td_1).
# SPECIFYING A SIMILAR NAME TO THE COLUMNS FOR EASIER MERGING LATER.
td_1 = pd.read_csv('Telcom_dataset1.csv')
td_1.columns = ['PRODUCT','VALUE','DATE_TIME','CELL_ON_SITE','DW_A_NUMBER','DW_B_NUMBER','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
td_1.loc[:7]
```

```
# CREATING A DATAFRAME FOR Telcom_dataset2(td_2).
# SPECIFYING A COMMON NAME FOR THE COLUMNS FOR EASIER MERGING LATER.
td_2 = pd.read_csv('Telcom_dataset2.csv')
td_2.columns = ['PRODUCT','VALUE','DATE_TIME','CELL_ON_SITE','DW_A_NUMBER','DW_B_NUMBER','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
td_2.loc[:7]
```

```
# CREATING A DATAFRAME FOR Telcom_dataset3(td_3).
# SPECIFYING A COMMON NAME FOR THE COLUMNS FOR EASIER MERGING LATER.
td_3 = pd.read_csv('Telcom_dataset3.csv')
td_3.columns = ['PRODUCT','VALUE','DATE_TIME','CELL_ON_SITE','DW_A_NUMBER','DW_B_NUMBER','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
td_3.loc[:7]
```

```
# SINCE THE THREE ABOVE DATASETS HAVE SIMILAR COLUMNS, WE ARE GOING TO MERGE THEM INTO ONE BIG DATASET
# AND ASSIGN IT THE DATAFRAME (td)
td = pd.concat([td_1,td_2,td_3], axis = 0)
td
```

```
# IN THE NEXT STEP WE CLEAN THE DATA BY DROPPING THE COLUMNS WE DON'T NEED IN td.
td_d = td.drop(columns = ['CELL_ON_SITE','DW_A_NUMBER','DW_B_NUMBER','COUNTRY_A','COUNTRY_B','SITE_ID'], axis=0)
td_d
```

```
# NEXT WE REMOVE THE DUPLICATE VALUES IN THE td_d
td_d.duplicated()
df = td_d.drop_duplicates()
df
```

```
# NOW THAT OUR TELCOM_DATASET IS CLEANED, WE CREATE A DATAFRAME FOR OUR cells_geo.csv DATASET(cg).
cg = pd.read_csv('cells_geo.csv', delimiter = ';')
cg.loc[:7]
```

```
# WE CLEAN OUR DATA BY DROPPING THE COLUMNS WE WON'T BE USING.
cg_d = cg.drop(columns = ['LOCALISATION','DECOUPZONE','ZONENAME','LONGITUDE','LATITUDE','SITE_CODE'], axis=0)
cg_d
```

```
# REMOVING ANY DUPLICATE VALUES
cg_d.duplicated()
df = cg_d.drop_duplicates()
df
```

```
# OUR NEXT STEP IS MERGING THE TWO DATAFRAMES WE HAVE PREPARED(td_d and cg_d).
# THIS JOINS THE Telcom_dataset(s) TO THE cells_geo DATASET.
master = td_d.merge(cg_d, on = ['CELL_ID'],how = 'outer')
master
```

```
# OUR FIRST STEP ALLLOWS US TO SEE WHICH SERVICE IS THE BIGGEST INCOME GENERATOR.
master.groupby(['PRODUCT']) [['VALUE']].sum()
# VOICE SERVICES IS OUR BIGGEST INCOME GENERATOR.
```

```
# OUR SECOND STEP IN ANALYSIS IS LOOKING AT WHICH CITIES WERE MOST USED FOR THE THREE DAYS.
master['VILLES'].value_counts()
# THE DATA SHOWS WHICH CITIES WERE MOST USED IN DESCENDING ORDER.
```
